```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import altair as alt
```

```python
df = pd.read_csv("../data/US_youtube_trending_data.csv")
df.head()
```

```python
time_fmt_df = df
time_fmt_df["trending_date"]= pd.to_datetime(df.trending_date)
time_fmt_df["publishedAt"]= pd.to_datetime(df.publishedAt)
time_fmt_df.head()
```

```python
tag_slt_df = pd.concat([time_fmt_df["trending_date"],time_fmt_df["tags"].str.split(pat='|',expand=False)], axis=1)
tag_slt_df
```

```python
tag_df = pd.DataFrame([[x] + [z] for x, y in zip(tag_slt_df.index,tag_slt_df.tags) for z in y])
tag_df = tag_df.merge(tag_slt_df,left_on=0,right_index=True)
tag_df.rename(columns={1:"tag_name",0:"frequency"},inplace=True)
tag_df.drop(["tags"],axis=1,inplace=True)
tag_df
```

```python
tag_y_df = tag_df[tag_df["trending_date"].dt.year.isin(np.arange(2021,2022))]
tag_ym_df = tag_y_df[tag_y_df["trending_date"].dt.month.isin(np.arange(10,11))]
tag_ym_df = tag_ym_df[["tag_name","frequency"]].groupby(["tag_name"]).count().sort_values(["frequency"],ascending=False).drop(index="[None]")
tag_ym_df.head(20)
```

```python
tag_ym_df.head(10).plot(kind='bar')
```

```python

lsyear = [2020,2021]
lsmonth = [8,9,10,11]
tdf = pd.DataFrame()
# time_tag_y_df = time_fmt_df[time_fmt_df["trending_date"].dt.year.isin(lsyear)]
# time_fmt_y_df.sort_values(["trending_date"],ascending=True)

for y1 in lsyear:
    for m1 in lsmonth:
        # format time column
        time_tag_y_df = time_fmt_df[time_fmt_df["trending_date"].dt.year.isin([y1])]
        time_tag_y_m_df = time_tag_y_df[time_tag_y_df["trending_date"].dt.month.isin([m1])]

        # tag_df: return a dataframe with tags and its frequency over a period of time
        tag_df = time_tag_y_m_df["tags"].str.split(pat='|',expand=True).stack().reset_index()
        tag_df = tag_df.rename(columns={0:"tag_name","level_0":"frequency_0","level_1":"frequency_1"})
        tag_df = tag_df[["tag_name","frequency_0"]].groupby(["tag_name"]).count().sort_values(["frequency_0"],ascending=False).drop(index="[None]")
        tag_df[["trending_date"]] = str(m1)+"/"+ str(y1)
        tag_df["trending_date"]= pd.to_datetime(tag_df.trending_date,format='%m/%Y',errors='coerce')
        tdf = pd.concat([tdf,tag_df.head(20)])


tdf = tdf.dropna()
tdf

# tdf.to_csv("tag_trending.csv")
# tdf.groupby([df.trending_date.dt.year,df.trending_date.dt.month])
```
