```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("US_youtube_trending_data.csv")
df.head()
```

```python
ndf = df.groupby(["video_id"],as_index=False)["trending_date"].count().groupby(["trending_date"],as_index=False)["video_id"].count().sort_values(["trending_date"],ascending=False).set_index(["trending_date"])
ndf.plot()
ndf
```

```python
time_fmt_df = df
time_fmt_df["trending_date"]= pd.to_datetime(df.trending_date)
time_fmt_df["publishedAt"]= pd.to_datetime(df.publishedAt)
time_fmt_y_df = time_fmt_df[time_fmt_df["trending_date"].dt.year.isin(np.arange(2020,2021))]
time_fmt_y_m_df = time_fmt_y_df[time_fmt_df["trending_date"].dt.month.isin(np.arange(1,13))]
time_fmt_y_m_df
```

```python
# tag_df: return a dataframe with tags and its frequency over a period of time
tag_df = time_fmt_y_m_df["tags"].str.split(pat='|',expand=True).stack().reset_index()
tag_df = tag_df.rename(columns={0:"tag_name","level_0":"frequency_0","level_1":"frequency_1"})
tdf = tag_df[["tag_name","frequency_0"]].groupby(["tag_name"]).count().sort_values(["frequency_0"],ascending=False).drop(index="[None]")
```

```python
tdf.head(10).plot(kind='bar')
```
